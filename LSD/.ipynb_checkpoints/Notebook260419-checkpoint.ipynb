### Finite Elements Assembly and Solution
#### Assembly 
Until here, no physics are involved. Now, a Finite Element code, which assembles elements according to the given topology is used. In-house finite element code AMfe. At fist, the geometry which connects the keypoints, e.g. a pipe or some kind of beam, 
has to be defined. 

```python
D = 0.2; d= 0.195
E = 2.1e11;nu = 0.3;G = E/(2*(1+nu)); rho = 7500
A   = np.pi*(D**2 - d**2)/4
I_y = np.pi*(D**4 - d**4)/64
I_z = np.pi*(D**4 - d**4)/64
J_x = np.pi*(D**4 - d**4)/32
J_x = np.pi*(D**4 - d**4)/64
#X3 = np.array([1E21, 1E21, 1E21]) # Be aware!
X3 = np.array([30, 15,15]) # Be aware!
```

These properties uniquely define a beam. Since a beam is defined between two points, the orientation around this axis must be defined. This is achieved by using a third point, variable X3, which defines the positive z-Axis of the profile by spanning a triangle. Note, that this point is commonly defined for all lines. It is important, that X3 is not part of any of these lines. Otherwise, this will not work because there is no triangle. 

```python
from amfe.mesh import Mesh
from amfe.component import StructuralComponent
from amfe.material import BeamMaterial, KirchhoffMaterial
from amfe.solver.translators import MechanicalSystem
```

In the finite element code, a mesh and elements can now be defined. THe AMfe code assembles the stiffness and mass matrix when the corresponding attribute is called 
Mfree = component.M(x0, dx0, t0)
Kfree = component.K(x0, dx0, t0)
the definition of the componet uses the mesh and the matriel, this includes the information of the beam. 

```python
mesh = Mesh(3)

for i, Keypoint in enumerate(HangarGeo.Gridpoints):
    mesh.add_node(Keypoint, i)
    
s = 'straight_line'
for i,LineElement in enumerate(HangarGeo.LineElements):
    mesh.add_element(s, np.array([LineElement.KP0id,LineElement.KP1id]), i)
    
component = StructuralComponent(mesh)
matBeam = BeamMaterial(E, G, rho, A, I_y, I_z, J_x, X3)
component.assign_material(matBeam, s, 'S', 'shape')
mapping = component._mapping

t0 = 0
x0 = np.zeros(mapping.no_of_dofs)
dx0 = x0.copy()
Mfree = component.M(x0, dx0, t0)
Kfree = component.K(x0, dx0, t0)
fig = pyplot.figure()
pyplot.subplot(121)
pyplot.spy(Mfree)
pyplot.subplot(122)
pyplot.spy(Mfree)
```

### Static Solution
For a non-trivial static solution, a load vector is required. Because an
>  object either remains at rest or continues to move at a constant velocity, unless acted upon by a force.

In order to define a load, the same procedure as for the boundary conditions is used. First their position, aka. node, then their direction, aka. degree of freedom is defined. For the present case, the applied force shall act on the four upmost nodes in negative z-direction.

```python
loadCoords = np.array([[18,0, 44],[42,0, 44],[18,30, 44],[42,30, 44]])
loadDof,_ = MKfix.Map.getByCoords(loadCoords,'y')

LoadExt = np.zeros(MKfix.Map.DofNumber)
LoadExt[loadDof] = -3
  
```

The assigment takes place on the unconstrained mapping. Therefore, the load vector must be limited to the free degreed of freedom in the last step. Just like the matrices. 

The static solution is defined as solution of the equaiton
\begin{equation}
Ku=f
\end{equation}
there are several possibilities to calcuate the solution $x$. One is by calculating the actual inverse of the problem:
\begin{equation}
K^{-1}K=KK^{-1}= I \rightarrow u = K^{-1}f
\end{equation}
Snce the calculation of the inverse is computationally costly, a different approach is commonly used to solve such inear equation, namely factorization. Factorization exploits two issues: 
- Splitting a matrix K in an upper and a lower triangular part $K=LU$ is much cheaper than calculating the inverse.
- Linear equation including triangluar matrices can be solved cheaply by substitution.

### Eigen Solution

We are interested in 
\begin{equation} \label{1}
(K-\omega^2M) \Phi =0
\end{equation}
The major problem is, that the schoolbook solution is not feasible. Which is stated by a rank deficient matrix $det(K-\omega^2M) =0$ and a the subsequent solution for a linear equation. Problem: calculating the determinant (symbolic), and solving for the roots of the resulting n order polynomial is not really possible. Additionally, when it comes to the Eigenbehavior of one is typically only interested in a certain frequency range, predominantly low frequencies. 

All iterative algorithms which aim to find a subset of the solutions to the eigenproblem  are based on the same priciple. 
The main question is how this can be done efficiently. 

#### Power Iteration
##### Basic routine
Basic eigenproblem
\begin{equation}
Ax =\lambda x \leftrightarrow M^{-1}K\Phi = \omega^2 \Phi \leftrightarrow K^{-1}M\Phi = \frac{1}{\omega^2} \Phi
\end{equation}
can be solved by iteratively multiplying $A$ on a initial guess $x_0$. This procedure will converge to the eigenvector belonging to the highest $\lambda$.

### Harmonic Solution
Besides the static and eigen solution, which is the the structures ongoing harmonic motion without any external load. Internal interia and static forces are perfectly balanced. Besides: the structures response to an external _harmonic_ excitation is of large interest. This procedure has string similarities with the predceeding eigen solution. The general equation of motion is 
\begin{equation}
M\ddot{x}+Kx = f
\end{equation}
in case of the eigen and harmonic solution, the frequency domain is required
\begin{equation}
KX-\omega^2M = F(\omega)
\end{equation}
For the eigensolution, the question was which harmonic motion is feasible for the structure without external excitation: $F(\omega)=0 \forall \omega$. In case of the harmonic solution, the right hand side is supposed to be nonzero. 

#### Direct Solution
Harmonic Solution comes down to calculating the Frequency Response Function, which is defined as:
\begin{equation}
Y(\omega) = (KX-\omega^2M)^{-1}
\end{equation}
It is a matrix which depends on the frequency. Above equation also referred to as the direct solution. Which means calcuating the inverse for all freuencies of interest. When $Y$ is known, the harmonic reponse of a certain harmonic load can be calculated. 


#### Modal Truncation
The FRF can also be approximated based on a subset of $k$ modes:
\begin{equation}
Y_{MT}(\omega)= \sum^{k}_{i=1} \frac{\Phi_i \Phi_i^T }{\omega^2_i - \omega^2}
\end{equation}
Again the neverending story: accuracy versus efficiency. 

#### Mode Acceleration Method
The Modal Truncation does not get the static repsonse, aka the value of $Y(0)$ right. Which is obviously $K^{-1}$. 
So we want (slightly different derivation compared to lecture)
\begin{equation} 
Y_{MT}(0) = \sum^{k}_{i=1} \frac{\Phi_i \Phi_i^T }{\omega^2_i} \\
Y_{MA}(0) \overset{!}{=} K^{-1} \\
\Rightarrow Y_{MA}(\omega) = Y_{MT}(\omega)-Y_{MT}(0)+K^{-1}=  \sum^{k}_{i=1}  \frac{\Phi_i \Phi_i^T }{\omega^2_i - \omega^2}+ \frac{\Phi_i \Phi_i^T }{\omega^2_i} + K^{-1}
\end{equation}

Several notes: the previously acquired results of the static solution clearly indicate that the evaluation of $Y_{MA}$ can be constly for all inputs.Preselect a load case and evaluate $K^{-1}F$ with the already known LU decomposition of $K$.
